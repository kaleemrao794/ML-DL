# **Artificial Neural Network**
* A neural network is a type of machine learning algorithm modeled\
  after the structure and
function of the human brain. It is composed of layers of interconnected\
"neurons," which
process and transmit information.
* Neural networks are trained using large amounts of data and an optimization\
  algorithm,
which adjusts the weights and biases of the connections between neurons. Once\
trained, the
neural network can make decisions or predictions based on new input data.
* Neural networks are commonly used in a variety of applications, including image\
  and speech
recognition, natural language processing, and even playing games. They are particularly\
well-suited for tasks that require the ability to learn and adapt to new data, as they can\
automatically extract useful features from raw data and improve their performance over\
time.

# **Difference between ML & DL**
  * **Machine learning (ML) and deep learning (DL)** are both subfields of\
  artificial intelligence (AI) that involve the use of algorithms to automatically\
  learn and improve from data, rather than being explicitly programmed. However,\
  there are some key differences between the two approaches:
    * **Depth of model:** Deep learning models are typically deeper, meaning they have more\
    layers, compared to machine learning models. This allows deep learning models to learn\
    more complex patterns in the data, but also requires more data and computational\
    resources to train.
    * **Input data:** Deep learning algorithms are well-suited for processing and learning from\
    complex, unstructured data, such as images, audio, and text. Machine learning\
    algorithms, on the other hand, can be used with structured or unstructured data, but\
    may not be as effective at learning from highly complex data.
  * Overall, deep learning is a specialized subset of machine learning that is particularly\
   effective at solving complex, data-intensive problems, but requires more resources and\
    expertise to implement.

* **Import Libraries**

* **Load the Fashion MNIST Dataset**

```python
fmnist = tf.keras.datasets.fashion_mnist
```

* Load the training and test split of the fashion MNIST

```python
(training_images, training_labels), (test_images, test_labels) = fmnist.load_data()
```

* **Show Training Image**

```python
import numpy as np
import matplotlib.pyplot as plt
index = 0
np.set_printoptions(linewidth=320)
print(f'LABEL:{training_labels[index]}')
print(f'nIMAGE PIXEL ARRAY:\n{training_images[index]}')
plt.imshow(training_images[index])
```

* **Normalize the pixel values**

```python
training_images = training_images/255.0
test_images = test_images/255.0
```

* **Build the Classification Model**

```python
model = tf.keras.models.Sequential([tf.keras.layers.Flatten(),
                                    tf.keras.layers.Dense(128, activation=tf.nn.relu),
                                    tf.keras.layers.Dense(10, activation=tf.nn.softmax)])
```

```python
# Declare sample input and convert to a tensor
inputs = np.array([[1.0, 3.0, 4.0, 2.0]])
inputs = tf.convert_to_tensor(inputs)

# Feed the inputs to a softmax activation function
outputs = tf.keras.activations.softmax(inputs)

# Get the sum of all values after the softmax
sum = tf.reduce_sum(outputs)
print(f'sum of outputs:{sum}')

# Get the index with highest values
prediction = np.argmax(outputs)
print(f'class with highest probability: {prediction}')
```

* **Compiling the Model**

```python
model.compile(optimizer=tf.optimizers.Adam(),
              loss='sparse_categorical_crossentropy',
              metrics= ['accuracy'])
```

* **Fitting the Model**

```python
model.fit(training_images, training_labels, epochs=5)
```

* **Evaluate the Model on unseen data**

```python
model.evaluate(test_images, test_labels)
```

### Exercise 8: 

Earlier when you trained for extra epochs you had an issue where your loss might change. It might have taken a bit of time for you to wait for the training to do that, and you might have thought 'wouldn't it be nice if I could stop the training when I reach a desired value?' -- i.e. 60% accuracy might be enough for you, and if you reach that after 3 epochs, why sit around waiting for it to finish a lot more epochs....So how would you fix that? Like any other program...you have callbacks! Let's see them in action...

* **With 60% Accuracy**

```python
class myCallback(tf.keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs={}):
    if(logs.get('accuracy') >= 0.6):
      print("\nReached 60% accuracy so cancelling training!")
      self.model.stop_training = True
callbacks = myCallback()
fmnist = tf.keras.datasets.fashion_mnist
(training_images, training_labels) ,  (test_images, test_labels) = fmnist.load_data()
training_images=training_images/255.0
test_images=test_images/255.0
model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(),
  tf.keras.layers.Dense(512, activation=tf.nn.relu),
  tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(training_images, training_labels, epochs=5, callbacks=[callbacks])
```

# **4-Gradient Descent**

* Gradient Descent is an optimization algorithm which is commonly used\
  to train machine learning models and neural networks.
* Help to achive local or global minima (minimum values of any model):
  * best learning rate
  * best loss rate
  * best cost function rate
  * weights to some extent

### **Dependency**
* It works only on convex function
  * With single steepest point
* It fails on non-convex function
  * With multiple steepest point (saddle wave)

### **Types**
* ***It has 3 types:***
  * **01-Batch GD**
  * **02-Stochastic GD**
  * **03-Mini-Batch GD**

* **01-Batch GD**
  * Updates model after all epochs
  * Requires more computational power and memory as well

* **02-Stochastic GD**
  * Updates model after each epochs
  * Requires less computational power and saves every memory

* **03-Mini-Batch GD**
  * Combination of both stochastic and batch
  * requires less computational power and less memory as well

### **Comparison of GD Types**
* **Batch GD** is best but uses more computation
* Mostly used **Stochastic GD** due to less computation requirement but\
  less effective than Batch GD
* **mini-batch GD** used seldomly

### **Challanges**
* If local minima reachs first than it is very difficult to attain global minima
* In saddle wave it is again difiicult to attain global minima

# **7-Activation functions**

* Activation functions are of following kinds:
  * **1- Binary Step Function**
    * This function is Threhold based function.i.e,
      * Neurons become activated after threhold point comes.
      * Neurons become deactivated below threshold point.
  * **2- Sigmoid function**
    * This function has a smooth values between 0 and 1 and hence have a smooth curve\
      rather than Binary Threshold function.
    * Therefore this function performs best for positive values.
  * **formula**
     $$f(x) = 1/1+e^{-x}$$
  * **Advantages:**
    * Use to predict probability as output.
    * Normalize the output value(between 0 & 1).
    * Differentiable and provides smooth gradient and prevent jumps.
  * **Disadvantages:**
    * Computational expensive.
    * Output not zero.
    * Has a vanishing gradient problem(Values above 1 are all equals to one).i.e,
      * Makes no change for very high or very low input which results in Neural\
        Network refuse to learn further, so use to predict low values.
  * **3- Tan-H (Hyperbolic Tangent)**
    * This activation function is very similar to Sigmoid function.
    * Difference with sigmoid is in between output value as it ranges from -1 to 1.
  * **formula**
     $$f(x) = \frac{e^{x}-e^{-x}}{e^{x}+e^{-x}}$$ 
  * **Advantages:**
    * It is zero centered.
    * Prediction is simple and based on threshold probability.
  * **Disadvantages:**
    * Vanish Gradient problem is also exist in it.
  * **4-Relu(Rectified Linear Unit)**
    * It is more efficient activation function as compared to sigmoid and tan-h as:
      * Output for positive input range from 0 to infinity.
      * It has a derivative function and allows for back propagation simultaneously\
        making computational efficient.
      * The main catch of ReLU is that, it not activate all neurons at same time.
      * Neurons only be deactivated if output of Linear transform is less than 0.
  * **formula**
  $$f(x) = max(0,x)$$
  * **Advantages:**
    * More efficient as use less computational power (Activate 1st than 2nd Neuron).
    * Accelerate Gradient Descent towards global minima.
  * **Disadvantages:**
    * When input is zero or negative its output becomes zero and hinders with back\
      propagation.
    * This problem is known as dying ReLU function.
    * Therefore usefull to use when there is positive values in the input.
  * **5-Leaky-ReLU**
    * This is extension of ReLU and use to resolve:
      * Used to solve dying ReLU.
      * It is Nothing but improved version of ReLU.
      * It adds a slight slope in negative range.
    * **formula**
     $$f(x) = max(0.1x,x)$$
    * **Advantages:**
      * Enable backpropagation even for negative value of ReLU.
      * Negative values of left side of grph comes out to be Non-zero.
    * **Disadvantages:**
      * Sust ReLU Problem
    * **6-Parametric ReLU**
      * This is another extension of ReLU and use to resolve:
        * It aims to resolve problem of gradient becoming zero for left half of axis.
        * This provides slope of negative part of function as an argument by\
          performing back-propagation, so most appropriate value of a is learnt.
    * **formula**
      $$f(x) = max(0.1x,x)$$
    * **Advantages:**
      * Parametric ReLU function is used when the Leaky ReLU function still fails\
        at solving problem of dead neuron, and the relavent information not successfuly\
        passed to next layer. 
    * **Disadvantages:**
      * Its limitation is that it may perform differently for different problems depending\
        upon the value of slope (parameter a).
    * **Softmax Activation**
      * It is used to perform following tasks:
        * It is used for multiple dimension.
        * Convert vector of number into vector of probablisties.
        * Used in case of Multi-class Classification.
        * Output of softmax are interparellel as probabilities of getting each class output\
          of sigmoid was in range of 0 to 1 can be thought as probability.
    * **formula**
       $$Softmax(Z_i) = \frac{exp(Z_i)}{\sum_j{exp(Z_j)}}$$
    * **Advantages:**
      * It can be used for Multi-Class classification.
      * Normalize the output of each class between 0 and 1 and divide by their sum,\
        giving the probability of input value being in a Speciafic class.
      * For Neural Network need to categorize input into nemerous categories.\
        Softmax is often employed exclusively for output layer.

#### **How to choose the best activation function?**  
  * **Activation functions for Hidden layers:**
    * Modern neural networks with common architectures, such as **MLP** and **CNN**\
      will make use of the **ReLU** activation function or extensions.
    * **Recurrent networks** still commonly use **Tan-H** or **Sigmoid** activation\ 
      function or both.
      * For example, the **LSTM** commonly use the **Sigmoid** for recurrent connection\
        and the **Tan-H** function for output.
  * **The Activation function for Hidden Layer** is typically chosen based on the type\
    of **neural networks architectures**.
      * **Multilayer Perceptron(MLP): ReLU activation function.**
      * **Convolutional Neural Network(CNN): ReLU activation function.**
      * **Recurrent Neural Network(RNN): Tan-H and\or Sigmoid activation function.**
  * **The Activation function for Output Layer** 
    * In output layer activation functions selected based on problem.i.e, Regression\
      or classification. For regression problems we would choose:
      * **Linear Activation function:**
        * The Linear activation is call **identity** or **no activation.** This is because\
        it does not change the weighted sum of input in any way and insert values directly.
    * For Classification or probaility we would choose following activation functions.\
      Remember Probability is also a classification problem.\
      In all cases of classification, your model will predict the probability(e.g. probability\
      that an example belongs to each class) that you can convert to a criss cross label by:
      * rounding(for **Sigmoid**)
      * argmax(for **Softmax**) 
      * **Logistic(Sigmoid) Activation function:**
        * For **two mutually exclusive**(Mutual exclusive classification is a type of\
          classification where each data point belongs to exactly one class. This means that\
          the classes are mutually exclusive, meaning that a data point cannot belong to more\
          than one class at the same time. It is **binary classification**). We would choose\
          **Sigmoid** Activation function for it. It has just one node.
        * For **two or more mutually inclusive**(Mutual inclusive classification is a type of\
          classification where each data point can belong to multiple classes at the same time.\
          This means that the classes are not mutually exclusive, and a data point can belong to\
          more than one class). It is a type called **multilabel classification.** In this case\
          output layer has one node for each class.  
      * **Softmax Activation function:**
        * For **more than two mutually exclusive (multiclass classification)** we would choose\
          **Softmax** Activation function. It have one node per class.

* Import libraries

```python
import keras
from keras.datasets import fashion_mnist
from keras.layers import Dense, Activation, Flatten, Conv2D, MaxPooling2D
from keras.models import Sequential
from keras.utils import to_categorical
import numpy as np
import matplotlib.pyplot as plt
```

* Loading Fashion MNIST Dataset 

```python
(train_X, train_y) , (test_X, test_y) = fashion_mnist.load_data()
```

* Shape of the dataset

```python
print('train_X.shape', train_X.shape)
print('train_y.shape', train_y.shape)
print('test_X.shape', test_X.shape)
print('test_y.shape', test_y.shape)
```

* Visualize data

```python
plt.imshow(train_X[0])
```

* Normalize the data

```python
train_X = train_X.astype('float32')/255
test_X = test_X.astype('float32')/255
```

* Reshaping the data

```python
train_X = train_X.reshape(-1, 28, 28, 1)
test_X = test_X.reshape(-1, 28, 28, 1)
```

* Changing our output to Categorical

```python
train_y_one_hot = to_categorical(train_y)
test_y_one_hot = to_categorical(test_y)
```

* **CNN**

```python
# object instance of Sequential Class
model = Sequential()
# Must define the input shape in the first layer of the Neural Network
# Convolutional layer
# Filters --> learned during training and provide brightness to the pixel on the basis of their # Kernal size --> height and width of 2D conv window or
# A number specifying both the height and width of the (square) convolutional window
model.add(Conv2D(64, (3,3), input_shape=(28, 28, 1)))
model.add(Activation('relu'))
# Maxpooling layer
model.add(MaxPooling2D(pool_size=(2,2)))
# Convolutional layer
model.add(Conv2D(64, (3,3)))
model.add(Activation('relu'))
# Maxpooling layer
model.add(MaxPooling2D(pool_size=(2,2)))
# Flatten to make our model 1D
model.add(Flatten())
# Fully connected layer
model.add(Dense(10))
model.add(Activation('softmax'))
# Summary
model.summary()
```

* Compiling Model

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

* Model Fitting

```python
model.fit(train_X, train_y_one_hot, epochs=5)
```

* Testing Our Model

```python
test_loss, test_acc = model.evaluate(test_X, test_y_one_hot)
print('Test_loss:', test_loss)
print('Test_acc:', test_acc)
```

* Prediction

```python
predictions = model.predict(test_X)
print(np.argmax(np.round(predictions[0])))
```

* Plotting Result

```python
plt.imshow(test_X[0].reshape(28,28), cmap=plt.cm.binary)
plt.show()
```

* Load Dataset

```python
remove_dir = os.path.join(train_dir, 'unsup')
shutil.rmtree(remove_dir)
```

```python
batch_size = 32
seed = 42
raw_train_ds = tf.keras.utils.text_dataset_from_directory(
    'D:/Python_Ka_Chilla_Data/proj1/aclImdb/train',
    batch_size=batch_size,
    validation_split=0.2,
    subset='training',
    seed=seed
)
```

```python
for text_batch, label_batch in raw_train_ds.take(1):
    for i in range(3):
        print('Review', text_batch.numpy()[i])
        print('Label', text_batch.numpy()[i])
```

```python
print('label 0 corresponds to', raw_train_ds.class_names[0])
print('label 1 corresponds to', raw_train_ds.class_names[1])
```

```python
raw_val_ds = tf.keras.utils.text_dataset_from_directory(
    'D:/Python_Ka_Chilla_Data/proj1/aclImdb/train',
    batch_size=batch_size,
    validation_split=0.2,
    subset='validation',
    seed=seed
)
```

```python
raw_test_ds = tf.keras.utils.text_dataset_from_directory(
    'D:/Python_Ka_Chilla_Data/proj1/aclImdb/test',
    batch_size=batch_size
)
```

* Prepare the Dataset for training
  * Standardization

```python
def custom_standardization(input_data):
    lowercase = tf.strings.lower(input_data)
    stripped_html = tf.strings.regex_replace(lowercase, '<br />', '  ')
    return tf.strings.regex_replace(stripped_html,
                                    '[%s]' % re.escape(string.punctuation),
                                    '')
```

```python
max_features = 10000
sequence_length = 250

vectorize_layer = layers.TextVectorization(
    standardize=custom_standardization,
    max_tokens=max_features,
    output_mode='int',
    output_sequence_length=sequence_length)
```

* Make a text-only dataset (without labels), then call adapt

```python
train_text = raw_train_ds.map(lambda x, y: x)
vectorize_layer.adapt(train_text)
```

```python
def vectorize_text(text, label):
  text = tf.expand_dims(text, -1)
  return vectorize_layer(text), label
```

```python
# retrieve a batch (of 32 reviews and labels) from the dataset
text_batch, label_batch = next(iter(raw_train_ds))
first_review, first_label = text_batch[0], label_batch[0]
print("Review", first_review)
print("Label", raw_train_ds.class_names[first_label])
print("Vectorized review", vectorize_text(first_review, first_label))
```

```python
print("1287 ---> ",vectorize_layer.get_vocabulary()[1287])
print(" 313 ---> ",vectorize_layer.get_vocabulary()[313])
print('Vocabulary size: {}'.format(len(vectorize_layer.get_vocabulary())))
```

```python
train_ds = raw_train_ds.map(vectorize_text)
val_ds = raw_val_ds.map(vectorize_text)
test_ds = raw_test_ds.map(vectorize_text)
```

* Configure the dataset for performance

```python
AUTOTUNE = tf.data.AUTOTUNE

train_ds = train_ds.cache().prefetch(buffer_size=AUTOTUNE)
val_ds = val_ds.cache().prefetch(buffer_size=AUTOTUNE)
test_ds = test_ds.cache().prefetch(buffer_size=AUTOTUNE)
```

* Create the model

```python
embedding_dim = 16
```

```python
model = tf.keras.Sequential([
  layers.Embedding(max_features + 1, embedding_dim),
  layers.Dropout(0.2),
  layers.GlobalAveragePooling1D(),
  layers.Dropout(0.2),
  layers.Dense(1)])

model.summary()
```

* Loss function and optimizer

```python
model.compile(loss=losses.BinaryCrossentropy(from_logits=True),
              optimizer='adam',
              metrics=tf.metrics.BinaryAccuracy(threshold=0.0))
```

* Train the model

```python
epochs = 10
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=epochs)
```

* Evaluate the model

```python
loss, accuracy = model.evaluate(test_ds)

print("Loss: ", loss)
print("Accuracy: ", accuracy)
```

* Create a plot of accuracy and loss over time

```python
history_dict = history.history
history_dict.keys()
```

```python
acc = history_dict['binary_accuracy']
val_acc = history_dict['val_binary_accuracy']
loss = history_dict['loss']
val_loss = history_dict['val_loss']
epochs = range(1, len(acc) + 1)

# "bo" is for "blue dot"
plt.plot(epochs, loss, 'bo', label='Training loss')
# b is for "solid blue line"
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and Validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and Validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.show()
```

* Export the model

```python
export_model = tf.keras.Sequential([
  vectorize_layer,
  model,
  layers.Activation('sigmoid')
])
export_model.compile(
    loss=losses.BinaryCrossentropy(from_logits=False), optimizer="adam", metrics=['accuracy']
)

# Test it with `raw_test_ds`, which yields raw strings
loss, accuracy = export_model.evaluate(raw_test_ds)
print(accuracy)
```

# **13-Preprocessing of NLP**

* In **natural language processing (NLP)**, it is often necessary to convert\
  raw text data into a more structured and numeric format that can be more\
  easily analyzed and processed by algorithms. There are several common\
  techniques that are used to accomplish this, including standardization,\
  tokenization, and vectorization.
  * **Standardization:**\
  Standardization refers to the process of cleaning and\
  preprocessing text data in order to make it more consistent and standardized.\
  This may involve tasks such as converting all text to lowercase, removing\
  punctuation, or removing stop words (common words that add little meaning\
  to a text, such as "the" or "and"). Standardization helps to ensure that the\
  text is in a consistent format and that the analysis is not biased by\
  variations in the text data.
  * **Tokenization:**\
  Tokenization is the process of dividing a text into smaller units called tokens.\
  These tokens can be individual words, phrases, or even subwords. Tokenization is\
  often used as a preprocessing step for natural language processing tasks, as it\
  allows the algorithms to more easily analyze and understand the meaning of the text.\
  * **Vectorization:**\
  Vectorization refers to the process of converting text data into numerical\
  vectors that can be used as input to machine learning algorithms. There are several\
  different methods for vectorizing text data, including term frequency-inverse document\
  frequency (TF-IDF), one-hot encoding, and word embeddings. Vectorization allows\
  algorithms to more easily work with text data by representing it in a numerical format that\
  they can understand.
